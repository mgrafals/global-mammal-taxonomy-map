The distribution can be made better if we bucket similar statuses into one. Therefore, this can lead to less filters with even distribution.

```python

def bucket_iucn_status(status):
    if status in ['EX', 'EW']:
        return 'Extinct'
    elif status in ['NT', 'CR', 'EN', 'VU']:
        return 'Threatened'
    elif pd.isna(status) or status == 'DD':
        return 'Data Deficient or Unknown'
    elif status == 'LC':
        return 'Least Concern'

df['iucn_bucket'] = df['iucnStatus'].apply(bucket_iucn_status)

df['iucn_bucket'].unique()
```

Got full form of the abbreviations from the data dictionary and bucketed accordingly.

```python
distinct_counts = df.groupby('iucn_bucket')['id'].nunique().reset_index()
distinct_counts.rename(columns={'id': 'distinct_id_count'}, inplace=True)
print(distinct_counts)
```

Will need to clean up this column and separate countries out

```python
def clean_country_distribution(df, column_name):
    print(df[column_name])
    df[column_name] = df[column_name].str.replace('?', '', regex=False)
    df[column_name] = df[column_name].str.split('|')
    return df

df_cleaned = clean_country_distribution(df, 'countryDistribution')
```

```python
all_countries = df_cleaned['countryDistribution'].explode().unique()
len(all_countries)
```

```python
df_exploded = df_cleaned.explode('countryDistribution')
df_exploded['countryDistribution'].unique()
```

```python
final_df = (
    df_exploded.groupby(['countryDistribution', 'iucn_bucket'])['id']
    .nunique()
    .reset_index(name='distinct_species_count')
)
final_df.head()
```

```python
def assign_buckets_with_ranges(final_df):
    def get_bucket_ranges(series):
        bins = pd.qcut(series.rank(method='first'), 5, retbins=True)[1]
        range_labels = [f"{int(bins[i])}-{int(bins[i+1]-1)}" for i in range(len(bins)-1)]
        return pd.qcut(series.rank(method='first'), 5, labels=range_labels)


    final_df['species_count_bucket'] = (
        final_df.groupby('iucn_bucket')['distinct_species_count']
        .transform(get_bucket_ranges)
    )
    return final_df


final_df = assign_buckets_with_ranges(final_df)


final_df.head()
#final_df.to_excel('/content/countryspeciesIUCNdistribution.xlsx', index=False)
```

```python
country_mapping = {
    'Bosnia & Herzegovina': 'Bosnia and Herz.',
    'Republic of the Congo': 'Congo',
    'Czech Republic': 'Czechia',
    'Dominican Republic': 'Dem. Rep. Congo',
    'Equatorial Guinea': 'Eq. Guinea',
    'Falkland Islands (Malvinas)': 'Falkland Is.',
    'Guinea-Bissau': 'Guinea-Bissau',
    'South Sudan': 'S. Sudan',
    'Solomon Islands': 'Solomon Is.',
    'United States': 'United States of America'
}

# Replace country names in final_df using the mapping
final_df['countryDistribution'] = final_df['countryDistribution'].replace(country_mapping)
final_df.head()
```

We need to join this excel with geojson file. Therefore, we will create a column that matches the geojson country name using fuzzy matching.

```python
import pandas as pd
import json
from rapidfuzz import process, fuzz

with open('/content/custom.geo.json', 'r') as f:
    geo_data = json.load(f)

# Extract country names from geo.json
geo_countries = [feature['properties']['name'] for feature in geo_data['features']]


# Normalize country names in both datasets
final_df['normalized_country'] = final_df['countryDistribution'].str.lower().str.replace(r'[^a-z\s]', '', regex=True)
geo_countries_normalized = [name.lower().replace("'", "").replace("-", " ") for name in geo_countries]

# Map countries using fuzzy matching
def map_country(country, geo_list):
    match = process.extractOne(country, geo_list, scorer=fuzz.token_sort_ratio)
    return match[0] if match and match[1] > 80 else None  # Threshold for matching

final_df['mapped_country'] = final_df['normalized_country'].apply(lambda x: map_country(x, geo_countries_normalized))
geo_df = pd.DataFrame({'geo_country': geo_countries, 'geo_country_normalized': geo_countries_normalized})
final_merged_df = final_df.merge(geo_df, left_on='mapped_country', right_on='geo_country_normalized', how='outer')

final_merged_df.to_excel('/content/countryspeciesIUCNmergeddataframedistribution.xlsx', index=False)
```
